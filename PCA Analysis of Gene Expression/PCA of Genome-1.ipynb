```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def ribo(genome):
    genome = ''.join(genome.split('\n'))
    amino_acids = []
    gene_list = []
    i_list = []
    n=1
    gene_mat = np.zeros([n,20])
    gene_num = 0

    # Find genes in genome: search for ATG start codon and stop codons
    for i in range(2,len(genome)-1):
        if i in i_list:
            continue
        elif ((genome[i-2]=='A') and (genome[i-1]=='T') and (genome[i]=='G')):
            for j in range(i,len(genome),3):
                if (((genome[j-2]=='T') and (genome[j-1]=='A') and (genome[j]=='G')) or 
                    ((genome[j-2]=='T') and (genome[j-1]=='A') and (genome[j]=='A')) or 
                    ((genome[j-2]=='T') and (genome[j-1]=='G') and (genome[j]=='A'))):
                    for index in range(i,j-2,3):
                        codon = [genome[index-2],genome[index-1],genome[index]]
                        gene_list.append(codon)
                    break
                
            #print(gene_list)      
            i_list = range(i,j)

            # Convert gene to amino acids
            for k in range(len(gene_list)):
                if (gene_list[k][0]=='A'):
                    if (gene_list[k][1]=='A'):
                        if (gene_list[k][2]=='T' or gene_list[k][2]=='C'):
                            amino_acids.append('Asn')
                        else:
                            amino_acids.append('Lys')
                    elif (gene_list[k][1]=='T'):
                        if (gene_list[k][2]!='G'):
                            amino_acids.append('Ile')
                        else:
                            amino_acids.append('Met')
                    elif (gene_list[k][1]=='C'):
                        amino_acids.append('Thr')
                    elif (gene_list[k][1]=='G'):
                        if (gene_list[k][2]=='T' or gene_list[k][2]=='C'):
                            amino_acids.append('Ser')
                        else:
                            amino_acids.append('Arg')
                elif (gene_list[k][0]=='T'):
                    if (gene_list[k][1]=='A'):
                        amino_acids.append('Tyr') 
                    elif (gene_list[k][1]=='T'):
                        if (gene_list[k][2]=='T' or gene_list[k][2]=='C'):
                            amino_acids.append('Phe')
                        else:
                            amino_acids.append('Leu')                   
                    elif (gene_list[k][1]=='C'):
                        amino_acids.append('Ser')   
                    elif (gene_list[k][1]=='G'):
                        if (gene_list[k][2]=='T' or gene_list[k][2]=='C'):
                            amino_acids.append('Cys')
                        else:
                            amino_acids.append('Trp')
                elif (gene_list[k][0]=='C'):
                    if (gene_list[k][1]=='A'):
                        if (gene_list[k][2]=='T' or gene_list[k][2]=='C'):
                            amino_acids.append('His')
                        else:
                            amino_acids.append('Gln')
                    elif (gene_list[k][1]=='T'):
                        amino_acids.append('Leu')
                    elif (gene_list[k][1]=='C'):
                        amino_acids.append('Pro')
                    elif (gene_list[k][1]=='G'):
                        amino_acids.append('Arg')
                elif (gene_list[k][0]=='G'):
                    if (gene_list[k][1]=='A'):
                        if (gene_list[k][2]=='T' or gene_list[k][2]=='C'):
                            amino_acids.append('Asp')
                        else:
                            amino_acids.append('Glu')
                    elif (gene_list[k][1]=='T'):
                        amino_acids.append('Val')
                    elif (gene_list[k][1]=='C'):
                        amino_acids.append('Ala')
                    elif (gene_list[k][1]=='G'):
                        amino_acids.append('Gly')          
            # Count number of each amino acid in gene
            for m in range(len(amino_acids)):
                if amino_acids[m] == 'Ala':
                    gene_mat[gene_num][0] = gene_mat[gene_num][0]+1
                elif amino_acids[m] == 'Arg':
                    gene_mat[gene_num][1] = gene_mat[gene_num][1]+1
                elif amino_acids[m] == 'Asn':
                    gene_mat[gene_num][2] = gene_mat[gene_num][2]+1
                elif amino_acids[m] == 'Asp':
                    gene_mat[gene_num][3] = gene_mat[gene_num][3]+1
                elif amino_acids[m] == 'Cys':
                    gene_mat[gene_num][4] = gene_mat[gene_num][4]+1
                elif amino_acids[m] == 'Gln':
                    gene_mat[gene_num][5] = gene_mat[gene_num][5]+1
                elif amino_acids[m] == 'Glu':
                    gene_mat[gene_num][6] = gene_mat[gene_num][6]+1
                elif amino_acids[m] == 'Gly':
                    gene_mat[gene_num][7] = gene_mat[gene_num][7]+1
                elif amino_acids[m] == 'His':
                    gene_mat[gene_num][8] = gene_mat[gene_num][8]+1
                elif amino_acids[m] == 'Ile':
                    gene_mat[gene_num][9] = gene_mat[gene_num][9]+1
                elif amino_acids[m] == 'Leu':
                    gene_mat[gene_num][10] = gene_mat[gene_num][10]+1
                elif amino_acids[m] == 'Lys':
                    gene_mat[gene_num][11] = gene_mat[gene_num][11]+1
                elif amino_acids[m] == 'Met':
                    gene_mat[gene_num][12] = gene_mat[gene_num][12]+1
                elif amino_acids[m] == 'Phe':
                    gene_mat[gene_num][13] = gene_mat[gene_num][13]+1
                elif amino_acids[m] == 'Pro':
                    gene_mat[gene_num][14] = gene_mat[gene_num][14]+1
                elif amino_acids[m] == 'Ser':
                    gene_mat[gene_num][15] = gene_mat[gene_num][15]+1
                elif amino_acids[m] == 'Thr':
                    gene_mat[gene_num][16] = gene_mat[gene_num][16]+1
                elif amino_acids[m] == 'Trp':
                    gene_mat[gene_num][17] = gene_mat[gene_num][17]+1
                elif amino_acids[m] == 'Tyr':
                    gene_mat[gene_num][18] = gene_mat[gene_num][18]+1
                else:
                    gene_mat[gene_num][19] = gene_mat[gene_num][19]+1
        
            # Change gene number for next search
            gene_num = gene_num + 1
            # Resize gene array
            n=n+1
            gene_mat.resize(n,20)
            gene_list = []
            amino_acids = []
            
    # Find proportions
    np.delete(gene_mat,n-1,axis=0)
    gene_mat=gene_mat/sum(gene_mat)
    
    return gene_mat

# Assumptions:
# Genome is 5' to 3'
# Every start codon is a gene
```

```python
genome = open('species1.txt','r')
genome = genome.read()
genome = ''.join(genome.split('\n'))
gene_mat = ribo(genome)
```

```python
def PCA(array):
    S = np.cov(array, rowvar=False)
    eigval, eigvect = np.linalg.eig(S)
    row, col = np.shape(eigvect)
    # Explained Ratio
    total = np.sum(eigval)
    ratio = eigval/total
    
    d = {'Ala': [eigval[0],0],
         'Arg': [eigval[1],1],
         'Asn': [eigval[2],2],
         'Asp': [eigval[3],3],
         'Cys': [eigval[4],4],
         'Gln': [eigval[5],5],
         'Glu': [eigval[6],6],
         'Gly': [eigval[7],7],
         'His': [eigval[8],8],
         'Ile': [eigval[9],9],
         'Leu': [eigval[10],10],
         'Lys': [eigval[11],11],
         'Met': [eigval[12],12],
         'Phe': [eigval[13],13],
         'Pro': [eigval[14],14],
         'Ser': [eigval[15],15],
         'Thr': [eigval[16],16],
         'Trp': [eigval[17],17],
         'Tyr': [eigval[18],18],
         'Val': [eigval[19],19]}
    df = pd.DataFrame(data=d)
    df = df.sort_values(by=0, axis = 1, ascending=False)
    df_arr = np.array(df)
    
    # Scree plot
    x = range(1,len(eigval)+1)
    plt.bar(x, ratio)
    plt.plot(x,ratio,c='red',label="Variance Ratio")
    plt.title("Scree Plot")
    plt.xlabel("Component Number")
    plt.ylabel("Explained Variance Ratio")
    plt.legend(loc="center right")
    plt.show()
    
    clr = np.array(["red","blue","green","purple","black"])
    
    scale = np.zeros(len(eigvect))
    for num in range(len(scale)):
        scale[num] = 1/(eigvect[0,int(df_arr[1,num])]**2+eigvect[1,int(df_arr[1,num])]**2)**0.5
        
    # Biplot
    PC1 = array @ eigvect[...,int(df_arr[1,0])]
    PC2 = array @ eigvect[...,int(df_arr[1,1])]
    plt.scatter(PC1, PC2, s=3, color='black')
    for i in range(3):
        plt.arrow(0,0,eigvect[0,int(df_arr[1,i])]*scale[i]/50,eigvect[1,int(df_arr[1,i])]*scale[i]/50,
                  width=0.0001,color=clr[i],label=df.columns[i])
    plt.title("Biplot")
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.legend(loc="best")
    plt.show()
    return df
```
